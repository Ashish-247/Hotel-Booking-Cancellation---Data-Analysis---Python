```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv("hotel_booking.csv")
dataset.head()
```

```python
dataset.columns
```

```python
dataset.drop(columns = ['name', 'email','phone-number', 'credit_card'], inplace = True)
```

```python
dataset.head(1)
```

```python
dataset.columns
```

```python
dataset["arrival_date_year"].unique()
```

```python
dataset.isnull().sum()
```

```python
(dataset.isnull().sum()/dataset.shape[0])*100
```

```python
dataset.fillna(0,inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.countplot(x = "hotel", data = dataset)
plt.title("No. of People visiting Hotels")
plt.show()
```

```python
dataset["is_canceled"].value_counts()
```

```python
plt.pie(dataset["is_canceled"].value_counts(), labels = ["Not canceled", "Canceled"], autopct="%0.2f%%")
plt.legend(["Not Canceled", "Canceled"], loc = 1)
plt.title("Percetage of People canceled Hotels")
plt.show()
```

```python
resort_data = dataset[dataset["hotel"]=="Resort Hotel"]
plt.pie(resort_data["is_canceled"].value_counts(), labels = ["Not canceled", "Canceled"], autopct="%0.2f%%")
plt.legend(["Not Canceled", "Canceled"], loc = 1)
plt.title("Percetage of People canceled Resort Hotels")
plt.show()
```

```python
city_data = dataset[dataset["hotel"]=="City Hotel"]
plt.pie(city_data["is_canceled"].value_counts(), labels = ["Not canceled", "Canceled"], autopct="%0.2f%%")
plt.legend(["Not Canceled", "Canceled"], loc = 1)
plt.title("Percetage of People canceled City Hotels")
plt.show()
```

```python
sns.countplot(x = "hotel", data = dataset, hue = "is_canceled")
plt.title("Cacelation and No_Cancelation status Hotel Wise")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x = "arrival_date_month", data = dataset)
plt.title("No. of People visiting Hotels per Month")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x = "arrival_date_month", data = dataset, hue = "hotel")
plt.title("No. of People visiting Hotels (CategoryWise) per Month")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x = "arrival_date_month", data = dataset, hue = "is_canceled")
plt.title("No. of People Canceling and Not Canceling Booking per Month")
plt.grid()
plt.show()
```

```python
resort_data = dataset[dataset["hotel"]=="Resort Hotel"]
plt.figure(figsize=(15,5))
sns.countplot(x = "arrival_date_month", data = resort_data, hue = "is_canceled")
plt.title("Cacelation and No_Cancelation status per Month for Resort Hotel")
plt.grid()
plt.show()
```

```python
city_data = dataset[dataset["hotel"]=="City Hotel"]
plt.figure(figsize=(15,5))
sns.countplot(x = "arrival_date_month", data = city_data, hue = "is_canceled")
plt.title("Cacelation and No_Cancelation status per Month for City Hotel")
plt.grid()
plt.show()
```

```python
top_10_country = dataset['country'].value_counts()[:10]
top_10_country.head(10)
```

```python
plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)
plt.title('10 Countries with Booking Percentge')
plt.figure(figsize=(6, 6))
plt.show()
```

```python
cancelled_data = dataset[dataset['is_canceled'] == 1]
top_10_cancelation_country = cancelled_data['country'].value_counts()[:10]
top_10_cancelation_country.head(10)
```

```python
cancelled_data = dataset[dataset['is_canceled'] == 1]
top_10_cancelation_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(9, 9))
plt.pie(top_10_cancelation_country, autopct='%.2f', labels=top_10_cancelation_country.index)
plt.title('Top 10 Countries with Booking Cancelation')
plt.show()
```
